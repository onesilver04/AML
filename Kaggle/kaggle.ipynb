```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('german_credit_data (1).csv')
data
```

## Data Cleansing

```python
percentage_missing=round(data.isnull().sum()/len(data)*100,2).sort_values(ascending = False)
print(percentage_missing)
```

```python
# 필요 없는 컬럼 제거
df=data.drop(columns=['Unnamed: 0'])
df.head()
```

```python
# 결측값 처리 - 최빈값
categoricals = ['Checking account','Saving accounts']

for num in categoricals :
    modus=df[num].mode().values[0]
    df[num] = df[num].fillna(modus)

df.head()
```

## 수치형, 범주형 변수 구분

```python
# Qualitative_cols = ['Sex', 'Job', 'Housing', 'Saving accounts', 'Checking account', 'Purpose', 'Risk']
# Quantitative_cols = ['Age', 'Credit amount', 'Duration']

# from sklearn.preprocessing import LabelEncoder
# import joblib

# sex_map = {
#     "male": 1,
#     "female": 0
# }

# risk_map = {
#     "good": 1,
#     "bad": 0
# }

# saving_account_map = {
#     "NaN": 0,
#     "little": 1,
#     "moderate": 2,
#     "quite rich": 3,
#     "rich": 4,
# }

# checking_account_map = {
#     "NaN": 0,
#     "little": 1,
#     "moderate": 2,
#     "rich": 3,
# }

# df["Sex"] = df["Sex"].map(sex_map)
# df["Risk"] = df["Risk"].map(risk_map)
# df["Saving accounts"] = df["Saving accounts"].map(saving_account_map)
# df["Checking account"] = df["Checking account"].map(checking_account_map)

# df = pd.get_dummies(
#     df,
#     columns=["Housing", "Purpose"],
#     drop_first=True
# )

# df.head()
```

```python
# 1️⃣ Risk만 인코딩
df["Risk"] = df["Risk"].astype(str).str.strip().str.lower()
df["Risk"] = (df["Risk"] == "good").astype(int)

# 2️⃣ 나머지 범주형 컬럼들
categorical_cols = ['Sex', 'Job', 'Housing', 'Saving accounts', 'Checking account', 'Purpose']

# 3️⃣ 전체 원핫 인코딩
df = pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=False  # RF는 drop_first 안 해도 됨
)

df.head()
```

## 타겟 분리

```python
target = "Risk"
X = df.drop(columns=[target])  # drop the target column
y = df[target]                  # target column

X
```

## RF baseline

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)

# 기본 RF (하이퍼파라미터 최소만: 재현성 위해 random_state만 고정)
rf = RandomForestClassifier(
    random_state=1,
    n_estimators=300,   # 기본(100)보다 조금 늘리면 보통 안정적
    n_jobs=-1
)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("Baseline RF Accuracy:", round(acc, 4))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))

# 확률이 있으면 AUC도 같이 확인(선택)
y_proba = rf.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_proba)
print("\nBaseline RF ROC-AUC:", round(auc, 4))
```

## 그리드서치 튜닝

```python
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=1, n_jobs=-1)

param_grid = {
    "n_estimators": [300, 600, 1000],
    "max_depth": [None, 6, 10, 14, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None],
    "bootstrap": [True, False]
}

grid = GridSearchCV(
    rf,
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

grid.fit(X_train, y_train)

best_rf = grid.best_estimator_

y_pred = best_rf.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("\n[Tuned RF Accuracy]:", round(acc, 4))
print("[Best Params]:", grid.best_params_)
```

## LightGBM

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=1
)

lgbm_base = lgb.LGBMClassifier(
    objective="binary",
    n_estimators=2000,          # early_stopping으로 자동으로 적절히 멈춤
    learning_rate=0.02,
    num_leaves=31,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=1,
    n_jobs=-1
)

lgbm_base.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    eval_metric="auc",
    callbacks=[lgb.early_stopping(stopping_rounds=200, verbose=False)]
)

pred = lgbm_base.predict(X_test)
proba = lgbm_base.predict_proba(X_test)[:, 1]

print("LGBM Baseline Accuracy:", round(accuracy_score(y_test, pred), 4))
print("LGBM Baseline ROC-AUC :", round(roc_auc_score(y_test, proba), 4))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, pred))
print("\nReport:\n", classification_report(y_test, pred, digits=4))
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

lgbm = lgb.LGBMClassifier(
    objective="binary",
    random_state=1,
    n_jobs=-1
)

param_dist = {
    "n_estimators": [1000, 2000, 4000],
    "learning_rate": [0.005, 0.01, 0.02, 0.05],
    "num_leaves": [15, 31, 63, 127],
    "max_depth": [-1, 3, 4, 5, 6, 8],
    "min_child_samples": [5, 10, 20, 30, 50],
    "subsample": [0.7, 0.8, 0.9, 1.0],
    "colsample_bytree": [0.7, 0.8, 0.9, 1.0],
    "reg_alpha": [0.0, 0.1, 0.5, 1.0],
    "reg_lambda": [0.0, 0.1, 0.5, 1.0],
}

search = RandomizedSearchCV(
    lgbm,
    param_distributions=param_dist,
    n_iter=60,
    scoring="roc_auc",   # ⭐ AUC로 튜닝
    cv=cv,
    random_state=1,
    n_jobs=-1,
    verbose=1,
    refit=True
)

search.fit(X_train, y_train)

best_lgbm = search.best_estimator_
print("Best CV AUC:", round(search.best_score_, 4))
print("Best params:", search.best_params_)

proba = best_lgbm.predict_proba(X_test)[:, 1]
print("Test AUC:", round(roc_auc_score(y_test, proba), 4))
```

```python
thresholds = np.linspace(0.05, 0.95, 181)
best_thr, best_acc = None, -1

for t in thresholds:
    pred_thr = (proba >= t).astype(int)
    acc = accuracy_score(y_test, pred_thr)
    if acc > best_acc:
        best_acc = acc
        best_thr = t

print("Best threshold:", round(best_thr, 3))
print("Best accuracy (threshold tuned):", round(best_acc, 4))

pred_final = (proba >= best_thr).astype(int)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, pred_final))
print("\nReport:\n", classification_report(y_test, pred_final, digits=4))
```
